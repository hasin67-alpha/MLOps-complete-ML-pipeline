## **Multiple model**

```python
# from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
import pickle
import os


def multiple_model(X_train, y_train, X_test, y_test):

    # logistic regression
    log_reg = LogisticRegression(max_iter=1000, random_state=42)

    # random forest
    rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)

    # xgboost
    xgb = XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, random_state=42, eval_metric='logloss')

    # Voting
    ensemble = VotingClassifier(
        estimators=[
            ('lr', log_reg),
            ('rf', rf),
            ('xgb', xgb)
        ],
        voting='soft'
    )


    models = {
        'Logistic Regression': log_reg,
        'Random Forest': rf,
        'XGBoost': xgb,
        'Ensemble (Voting)': ensemble
    }




    best_accuracy = 0
    best_model_name = ""
    trained_models= {}

    print("--- Model Training and Evaluation --- \n")

    for name, model in models.items():


        # model
        model.fit(X_train, y_train)

        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)

        print(f"{name} Accuracy: {acc:.4f}")

        print("Classification Report: \n", classification_report(y_test, y_pred))


        trained_models[name] = model
        if acc > best_accuracy:
            best_accuracy = acc
            best_model_name = name

    print(f"\n🏆 Best Model: {best_model_name} with Accuracy: {best_accuracy:.4f}")


    os.makedirs("model", exist_ok=True)

    # Extract your highest performing model from the tracking dictionary
    best_model_object = trained_models[best_model_name]

    # Correctly pass BOTH the model object and the file handler to pickle
    with open("model/model.pkl", "wb") as file:
        pickle.dump(best_model_object, file)
        print(f"📦 Best Model ({best_model_name}) saved successfully as 'model/model.pkl'!")

    # return trained_models
```

## **X and y are imbalance**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10, stratify=y)
multiple_model(X_train, y_train, X_test, y_test)
```
